```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
import copy
def choice_sort(matrix):
    for i in range(0, len(matrix) - 1):
        minimum = i
        for j in range(i + 1, len(matrix)):
            if matrix[j] < matrix[minimum]:
                minimum = j
        matrix[i], matrix[minimum] = matrix[minimum], matrix[i]
    return(matrix)

matrix_for_choice_sort = copy.deepcopy(matrix)

start_time = timeit.default_timer()

for i in range(int(user_m)):
 print(choice_sort(matrix_for_choice_sort[i]))

print("Time of sort", timeit.default_timer() - start_time)
```

```python
# Сортировка вставкой.
import timeit
import copy
def insert_sort(matrix):
    for i in range(len(matrix)):
        cursor = matrix[i]
        pos = i
        while pos > 0 and matrix[pos - 1] > cursor:
            matrix[pos] = matrix[pos - 1]
            pos = pos - 1
        matrix[pos] = cursor
    return (matrix)

start_time = timeit.default_timer()
matrix_for_insert_sort = copy.deepcopy(matrix)
for i in range(user_m):
    print(insert_sort(matrix[i]))

print("Time of sort", timeit.default_timer() - start_time)
```

```python
# Сортировка обменом. || Сортировка пузырьком.

def bubble_sort(matrix):
    def swap(i, j):
        matrix[i], matrix[j] = matrix[j], matrix[i]

    n = len(matrix)
    swapped = True
    
    x = -1
    while swapped:
        swapped = False
        x = x + 1
        for i in range(1, n-x):
            if matrix[i - 1] > matrix[i]:
                swap(i - 1, i)
                swapped = True
    return matrix

matrix_for_bubble_sort = copy.deepcopy(matrix)
start_time = timeit.default_timer()

for i in range(int(user_m)):
    print(bubble_sort(matrix_for_bubble_sort[i]))
                
print("Time of sort", timeit.default_timer() - start_time)
```

```python
# Сортировка Шелла.
import math

def shell_sort(matrix):
    interval = len(matrix) // 2
    while interval:
        for i, el in enumerate(matrix):
            while i >= interval and matrix[i - interval] > el:
                matrix[i] = matrix[i - interval]
                i -= interval
            matrix[i] = el
        interval = 1 if interval == 2 else int(interval * 5.0 / 11)
    return matrix

start_time = timeit.default_timer()

matrix_for_shell_sort = copy.deepcopy(matrix)
for i in range(int(user_m)):
    print(shell_sort(matrix_for_shell_sort[i]))
print("Time of sort", timeit.default_timer() - start_time)
```

```python
# Быстрая сортировка.

def quick_sort(matrix):
    if len(matrix)> 1:
        pillar=matrix.pop()
        greater_list, equal_list, smaller_list = [], [pillar], []
        for item in matrix:
            if item == matrix:
                equal_list.append(item)
            elif item > pillar:
                greater_list.append(item)
            else:
                smaller_list.append(item)
        return (quick_sort(smaller_list) + equal_list + quick_sort(greater_list))
    else:
        return matrix
    
matrix_for_quick_sort = copy.deepcopy(matrix)
start_time = timeit.default_timer()
for i in range(int(user_m)):
    print(quick_sort(matrix_for_quick_sort[i]))
print("Time of sort", timeit.default_timer() - start_time)
    
```

```python
# Турнирная сортировка.

def tournament_sort(a: list[int]) -> list[int]:
    a_len: int = len(a)
    heap = PriorityHeap()
    winners: list[int] = []
    losers: list[int] = []

    while heap.size < heap.max_size and heap.size < a_len:
        elem = a[heap.size]
        if winners and winners[-1] > elem:
            losers.append(elem)
        else:
            heap.max_heap_insert(a[heap.size])

    for i in range(heap.size, a_len):
        elem: int = a[i]

        if winners and winners[-1] > elem:
            losers.append(elem)
        else: 
            max_elem: int = heap.pop_max()

            if not winners:
                winners.append(max_elem)
            elif winners[-1] <= max_elem:
                winners.append(max_elem)
            else:
                losers.append(max_elem)

            heap.max_heap_insert(elem)
    
    while heap.size > 0:
        max_elem: int = heap.pop_max()

        if not winners:
            winners.append(max_elem)
        elif winners[-1] <= max_elem:
            winners.append(max_elem)
        else: 
            losers.append(max_elem)
    
    if not losers:
        return winners
    else:
        winners2: list[int] = tournament_sort(losers)

        i: int = 0
        j: int = 0
        winners1_len: int = len(winners)
        winners2_len: int = len(winners2)

        winners_res: list[int] = []

        while i < winners1_len and j < winners2_len:
            if winners[i] < winners2[j]:
                winners_res.append(winners[i])
                i += 1
            else:
                winners_res.append(winners2[j])
                j += 1
        while i < winners1_len:
            winners_res.append(winners[i])
            i += 1
        while j < winners2_len:
            winners_res.append(winners2[j])
            j += 1
        return winners_res

class PriorityHeap:
    def __init__(self) -> None:
        self.cache: list[int] = []
        self.size: int = 0
        self.max_size = 32

    def pop_max(self) -> any: ###
        res = self.cache[0]
        self.size -= 1
        self.cache[0] = self.cache[self.size]
        self.cache.pop(self.size)
        self.iter_heapify(0)
        return res

    def max_heap_insert(self, key: int):
        self.cache.append(-1000000000000000)
        self.insert_key(key, self.size)       #iter_key
        self.size += 1
        
    def insert_key(self, value: int, i: int):
        self.cache[i]  = value
        parent = self.parent(i)
        while i > 0 and self.cache[parent] > self.cache[i]:
            self.cache[parent], self.cache[i] = self.cache[i], self.cache[parent]
            i = parent
            parent = self.parent(i)

    def parent(self, i: int) -> int:
        res = (i >> 1) - 1
        if res < 0:
            return 0
        else:
            return res

    def iter_heapify(self, i: int) -> None:
        while True:
            l: int = (i << 1) + 1
            r: int = (i << 1) + 2
            largest: int = i

            if l < self.size and self.cache[l] < self.cache[largest]:
                largest = l

            if r < self.size and self.cache[r] < self.cache[largest]:
                largest = r

            if largest ==i:
                break
            else:
                self.cache[i], self.cache[largest] = self.cache[largest], self.cache[i]
                i = largest


matrix_for_tournament_sort = matrix
start_time = timeit.default_timer()
for i in range(int(user_m)):
    print(tournament_sort(matrix_for_tournament_sort[i]))
print("Time of sort", timeit.default_timer() - start_time)
```
