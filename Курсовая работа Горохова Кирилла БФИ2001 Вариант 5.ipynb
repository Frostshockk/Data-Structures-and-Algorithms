–í —Ç—Ä–µ—Ç—å–µ–º –ø—Ä–∏–º–µ—Ä–µ –Ω–∞ —Ä—ã–Ω–∫–µ –µ—Å—Ç—å —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω —Å—É–≤–µ–Ω–∏—Ä —Å—Ç–æ–∏–º–æ—Å—Ç—å—é 8 —Ñ—É–Ω—Ç–æ–≤, –°–∞—Ö–∏—Ä –Ω–µ –º–æ–∂–µ—Ç –µ–≥–æ –∫—É–ø–∏—Ç—å.

```python
import math
import timeit
import numpy as np
import heapq
import math
```

```python
def BinarySearch(Number_of_Souvenirs:int,Budget:int,Price_of_souv:list ): # –ë–∏–Ω–∞—Ä–Ω—ã–π –ø–æ–∏—Å–∫ –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –ª—É—á—à–µ–≥–æ –≤–∞—Ä–∏–∞–Ω—Ç–∞ –ø–æ–∫—É–ø–∫–∏
    first = 0 #–ü–µ—Ä–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ =0
    last = Number_of_Souvenirs #–ü–æ—Å–ª–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–µ = —á–∏—Å–ª—É —Å—É–≤–µ–Ω–∏—Ä–æ–≤
    while (first <= last): 
        mid = (first+last)//2 #–ò—â–µ–º —Å–µ—Ä–µ–¥–∏–Ω—É 
        if price(mid, Price_of_souv) == Budget: #–ï—Å–ª–∏ —Ü–µ–Ω–∞ —Ä–∞–≤–Ω–∞ –±—é–¥–∂–µ—Ç–∞, —Ç–æ –≤–æ–∑–≤—Ä–∞—â–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ –≤ —Å–µ—Ä–µ–¥–∏–Ω–µ
            return mid
        elif price(mid, Price_of_souv) < Budget: #–ï—Å–ª–∏ –Ω–∏–∂–µ –±—é–¥–∂–µ—Ç–∞, –±–µ—Ä—ë–º —Å–ª–µ–¥—É—é—â–∏–π —ç–ª–µ–º–µ–Ω—Ç, –∏–Ω–∞—á–µ –ø—Ä–µ–¥—ã–¥—É—â–∏–π
            first = mid + 1
        else:
            last = mid - 1
    return first - 1

def price(Budget: int, Price_of_souv: int): # –†–∞—Å—á—ë—Ç –Ω–æ–≤–æ–π —Å—Ç–æ–∏–º–æ—Å—Ç–∏
    ar = []
    for i in range(len(Price_of_souv)):
        p = Price_of_souv[i] + (i+1)*Budget #–†–∞—Å—á—ë—Ç –ø–æ —Ü–µ–Ω—ã –ø–æ —Ñ–æ—Ä–º—É–ª–µ
        ar.append(p) #–î–æ–±–∞–≤–ª—è–µ–º –≤ –º–∞—Å—Å–∏–≤
    ar.sort() #–°–æ—Ä—Ç–∏—Ä—É–µ–º
    return sum(ar[:Budget]) #–≤–æ–∑–≤—Ä–∞—â–∞–µ–º —Å—É–º–º—É —Ü–µ–Ω –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –±—é–¥–∂–µ—Ç–∞

def input_values():
    print("–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—É–≤–µ–Ω–∏—Ä–æ–≤ –∏ –±—é–¥–∂–µ—Ç –°–∞—Ö–∏—Ä–∞: ")
    Number_of_Souvenirs, Budget=map(int, input().split()) #–í–≤–æ–¥ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —Å—É–≤–µ–Ω–∏—Ä–æ–≤ –∏ –±—é–¥–∂–µ—Ç–∞ –°–∞—Ö–∏—Ä–∞
    check_type_int(Number_of_Souvenirs) #–ü—Ä–æ–≤–µ—Ä–∫–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —Å—É–≤–µ–Ω–∏—Ä–æ–≤ –Ω–∞ —Ç–∏–ø int 
    check_type_int(Budget)
    check_value(Number_of_Souvenirs, 1, 10, 5) #–ü—Ä–æ–≤–µ—Ä–∫–∞ –≤—ã—Ö–æ–¥–∞ –∑–∞ –≥—Ä–∞–Ω–∏—Ü—ã –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–π
    check_value(Budget, 1, 10, 9)
    print("–í–≤–µ–¥–∏—Ç–µ —Ü–µ–Ω—ã –Ω–∞ —Å—É–≤–µ–Ω–∏—Ä—ã: ")
    Price_of_souv = list(map(int, input().split())) # –í–≤–æ–¥ —Ü–µ–Ω –Ω–∞ —Å—É–≤–µ–Ω–∏—Ä—ã
    check_type_int_for_elems_in_list(Price_of_souv) # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ –≤–≤–æ–¥ int –∑–Ω–∞—á–µ–Ω–∏–π —ç–ª–µ–º–µ–Ω—Ç–æ–≤ list
    check_value_in_list(Price_of_souv, 1, 10, 9) # –ü—Ä–æ–≤–µ—Ä–∫–∞ –≤—ã—Ö–æ–¥–∞ 
    return  Number_of_Souvenirs, Budget, Price_of_souv

def check_type_int (n: int):
    if not isinstance(n, int): # –ü—Ä–æ–≤–µ—Ä–∫–∞ n –Ω–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ —Ç–∏–ø—É –¥–∞–Ω–Ω—ã—Ö int
        raise TypeError("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—É–≤–µ–Ω–∏—Ä–æ–≤ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å —Ü–µ–ª—ã–º —á–∏—Å–ª–æ–º") # –ü–æ–¥–Ω–∏–º–∞–µ—Ç –∏—Å–∫–ª—é—á–µ–Ω–∏–µ TypeError

def check_value(n: int, left_edge: int, right_edge: int, power_for_right_edge: int):
    if not (left_edge <= n <= np.power(right_edge, power_for_right_edge)): # –ü—Ä–æ–≤–µ—Ä–∫–∞ n –Ω–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ –≤—Ö–æ–∂–¥–µ–Ω–∏—è –≤ –≥—Ä–∞–Ω–∏—Ü—ã –¥–æ–ø—É—Å—Ç–∏–º—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
        raise ValueError("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—É–≤–µ–Ω–∏—Ä–æ–≤ –¥–æ–ª–∂–Ω–æ –ø—Ä–∏–Ω–∏–º–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ –æ—Ç 1 –¥–æ 10^5") # –ü–æ–¥–Ω–∏–º–∞–µ—Ç –∏—Å–∫–ª—é—á–µ–Ω–∏–µ ValueError

def check_type_int_for_elems_in_list(n: list):
    for elems in range(len(n)):
        if not isinstance(n[elems], int):
            raise TypeError("–ö–∞–∂–¥–∞—è —Ü–µ–Ω–∞ —Å—É–≤–µ–Ω–∏—Ä–∞ –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å —Ü–µ–ª—ã–º —á–∏—Å–ª–æ–º")

def check_value_in_list(n: list,left_edge: int, right_edge: int, power_for_right_edge: int):
    for elems in range(len(n)):
        if not (left_edge <= n[elems] <= np.power(right_edge,power_for_right_edge)):
            raise ValueError("–í—Å—è —Å—Ç–æ–∏–º–æ—Å—Ç—å —Å—É–≤–µ–Ω–∏—Ä–æ–≤ –¥–æ–ª–∂–Ω–∞ –ø—Ä–∏–Ω–∏–º–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ –æ—Ç 1 –¥–æ 10^9")

if __name__ == '__main__':
    start_time = timeit.default_timer() # –°–æ–∑–¥–∞–Ω–∏–µ —Ç–∞–π–º–µ—Ä–∞
    Number_of_Souvenirs, Budget, Price_of_souv =input_values() 
    temp_solve = BinarySearch(Number_of_Souvenirs, Budget, Price_of_souv) # –õ—É—á—à–∏–π –≤–∞—Ä–∏–∞–Ω—Ç –ø–æ–∫—É–ø–∫–∏
    print("–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: ")
    print(f"{temp_solve} {price(temp_solve,Price_of_souv)}") # –†–∞—Å—á—ë—Ç —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –ª—É—á—à–µ–≥–æ –≤–∞—Ä–∏–∞–Ω—Ç–∞ –ø–æ–∫—É–ø–∫–∏
    print("–í—Ä–µ–º—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –ø—Ä–æ–≥—Ä–∞–º–º—ã(sec) :", (timeit.default_timer() - start_time)) #–í—ã—á–∏—Å–ª–µ–Ω–∏–µ –∏ –≤—ã–≤–æ–¥ –æ–±—â–µ–≥–æ –∑–∞—Ç—Ä–∞—á–µ–Ω–Ω–æ–≥–æ –≤—Ä–µ–º–µ–Ω–∏
```

```python

def check_type_int (n: int):
    if not isinstance(n, int): # –ü—Ä–æ–≤–µ—Ä–∫–∞ n –Ω–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ —Ç–∏–ø—É –¥–∞–Ω–Ω—ã—Ö int
        raise TypeError("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–æ–∫—Å—ë—Ä–æ–≤ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å —Ü–µ–ª—ã–º —á–∏—Å–ª–æ–º") # –ü–æ–¥–Ω–∏–º–∞–µ—Ç –∏—Å–∫–ª—é—á–µ–Ω–∏–µ TypeError

def check_value(n: int, left_edge: int, right_edge: int):
    if not (left_edge <= n <= right_edge): # –ü—Ä–æ–≤–µ—Ä–∫–∞ n –Ω–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ –≤—Ö–æ–∂–¥–µ–Ω–∏—è –≤ –≥—Ä–∞–Ω–∏—Ü—ã –¥–æ–ø—É—Å—Ç–∏–º—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
        raise ValueError("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–æ–∫—Å—ë—Ä–æ–≤ –¥–æ–ª–∂–Ω–æ –ø—Ä–∏–Ω–∏–º–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ –æ—Ç 1 –¥–æ 150000") # –ü–æ–¥–Ω–∏–º–∞–µ—Ç –∏—Å–∫–ª—é—á–µ–Ω–∏–µ ValueError

def check_type_int_for_elems_in_list(n: list):
    for elems in range(len(n)):
        if not isinstance(n[elems], int): #–ü—Ä–æ–≤–µ—Ä–∫–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å–ø–∏—Å–∫–∞ –Ω–∞ —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö Int
            raise TypeError("–í–µ—Å –±–æ–∫—Å—ë—Ä–∞ –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å —Ü–µ–ª—ã–º —á–∏—Å–ª–æ–º") # –í—ã–∑—ã–≤–∞–µ—Ç –∏—Å–∫–ª—é—á–µ–Ω–∏–µ TypeError

def check_value_in_list(n: list, left_edge: int, right_edge: int):
    for elems in range(len(n)):
        if not (left_edge <= n[elems] <= right_edge): #–ü—Ä–æ–≤–µ—Ä–∫–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å–ø–∏—Å–∫–∞ –Ω–∞ –≤—ã—Ö–æ–¥ –∏–∑ –≥—Ä–∞–Ω–∏—Ü –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–π
            raise ValueError("–í–µ—Å –±–æ–∫—Å—ë—Ä–∞ –¥–æ–ª–∂–µ–Ω –ø—Ä–∏–Ω–∏–º–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ –æ—Ç 1 –¥–æ 150000") #–í—ã–∑–æ–≤ –∏—Å–∫–ª—é—á–µ–Ω–∏—è ValueError

def check_types_and_values_for_int_list(n: int, x: list):
    check_type_int(n) # –ü—Ä–æ–≤–µ—Ä–∫–∞ —á–∏—Å–ª–∞ n –Ω–∞ —Ç–∏–ø int
    check_value(n, 1, 150000) # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ –≤—ã—Ö–æ–¥ –∏–∑ –≥—Ä–∞–Ω–∏—Ü—ã –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–π
    check_type_int_for_elems_in_list(x) #–ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ —Ç–∏–ø —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ —Å–ø–∏—Å–∫–µ
    check_value_in_list(x, 1, 150000) #–ü—Ä–æ–≤–µ—Ä–∫–∞ –∑–Ω–∞—á–µ–Ω–∏–π —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å–ø–∏—Å–∫–∞ –Ω–∞ –≤—ã—Ö–æ–¥ –∏–∑ –≥—Ä–∞–Ω–∏—Ü –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–π

def solution(z: int, list_weight: list):
    last = 0 # —Å—á—ë—Ç—á–∏–∫ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–µ–ª–æ–≤–µ–∫
    for elems in list_weight:
        if last > elems: # –µ—Å–ª–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–µ–ª–æ–≤–µ–∫ –≤ –∫–æ–º–∞–Ω–¥–µ –±–æ–ª—å—à–µ, —á–µ–º –≤ —Å–ø–∏—Å–∫–µ –≤—ã—á–∏—Ç–∞–µ–º 1
            z-=1
        else:
            last = max(elems - 1, last + 1) # –≤—ã–±–∏—Ä–∞–µ–º –≤–∞—Ä–∏–∞–Ω—Ç–∞ —Å –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º —á–µ–ª–æ–≤–µ–∫ –≤ –∫–æ–º–∞–Ω–¥–µ
    return z

start_time = timeit.default_timer()
print("–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: \n –í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–æ–∫—Å—ë—Ä–æ–≤")
z = int(input()) # –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–æ–∫—Å—ë—Ä–æ–≤
list_weight = list(sorted(map(int, input().split()))) #–≤–µ—Å
check_types_and_values_for_int_list(z,list_weight)
solution(z,list_weight)
print("–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: \n –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–µ–ª–æ–≤–µ–∫ –≤ –∫–æ–º–∞–Ω–¥–µ —Ä–∞–≤–Ω–æ:")
print(z)
print("–í—Ä–µ–º—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –ø—Ä–æ–≥—Ä–∞–º–º—ã(sec) :", (timeit.default_timer() - start_time)) #–û–∫–æ–Ω—á–∞–Ω–∏–µ —Ç–∞–π–º–µ—Ä–∞ –∏ –≤—ã–≤–æ–¥ –ø–æ—Ç—Ä–∞—á–µ–Ω–Ω–æ–≥–æ –Ω–∞ —Ä–µ—à–µ–Ω–∏–µ –≤—Ä–µ–º–µ–Ω–∏
```

```python
class Graph:

    def __init__(self):  # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö –∏ —Ñ—É–Ω–∫—Ü–∏–∏ load
        self.numVertices = None  # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –≤–µ—Ä—à–∏–Ω
        self.numEdges = None  # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –∫—Ä–∞—ë–≤
        self.neighbors = None  # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è —Å–æ—Å–µ–¥–Ω–∏—Ö –≤–µ—Ä—à–∏–Ω
        self.loadGraph()

    def loadGraph(self):  # –§—É–Ω–∫—Ü–∏—è —Å–æ–∑–¥–∞–Ω–∏—è –≥—Ä–∞—Ñ–∞
        print("–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: \n –í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–µ—Ä—à–∏–Ω –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä—ë–±–µ—Ä")
        self.numVertices, self.numEdges = map(int, input().split())  # –í–≤–æ–¥ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤–µ—Ä—à–∏–Ω –∏ —Ä—ë–±–µ—Ä
        if not isinstance(self.numVertices, int):  # –ü—Ä–æ–≤–µ—Ä–∫–∞ –≤–µ—Ä—à–∏–Ω –Ω–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ —Ç–∏–ø—É –¥–∞–Ω–Ω—ã—Ö int
            raise TypeError("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–µ—Ä—à–∏–Ω –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å —Ü–µ–ª—ã–º —á–∏—Å–ª–æ–º")  # –ü–æ–¥–Ω–∏–º–∞–µ—Ç –∏—Å–∫–ª—é—á–µ–Ω–∏–µ TypeError
        if not isinstance(self.numEdges, int):
            raise TypeError("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä—ë–±–µ—Ä –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å —Ü–µ–ª—ã–º —á–∏—Å–ª–æ–º")
        if not (2 <= self.numVertices <= 100000):  # –ü—Ä–æ–≤–µ—Ä–∫–∞ –≤–µ—Ä—à–∏–Ω –Ω–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ –≤—Ö–æ–∂–¥–µ–Ω–∏—è –≤ –≥—Ä–∞–Ω–∏—Ü—ã –¥–æ–ø—É—Å—Ç–∏–º—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
            raise ValueError("–ó–Ω–∞—á–µ–Ω–∏–µ –≤–µ—Ä—à–∏–Ω—ã –¥–æ–ª–∂–Ω–æ –ø—Ä–∏–Ω–∏–º–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ –æ—Ç 1 –¥–æ 100000")  # –ü–æ–¥–Ω–∏–º–∞–µ—Ç –∏—Å–∫–ª—é—á–µ–Ω–∏–µ ValueError
        if not (2 <= self.numEdges <= 100000):
            raise ValueError("–ó–Ω–∞—á–µ–Ω–∏–µ —Ä—ë–±–µ—Ä –¥–æ–ª–∂–Ω–æ –ø—Ä–∏–Ω–∏–º–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ –æ—Ç 1 –¥–æ 100000")
        self.numVertices += 1  # –°—á—ë—Ç—á–∏–∫
        self.neighbors = tuple(list() for vertex in range(self.numVertices))  # –ó–∞–ø–æ–ª–Ω–µ–Ω–∏–µ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ —á–µ—Ä–µ–ø–∞—à–∫—É
        self.groups = (set(), set())

        for edge in range(self.numEdges):
            vertex1, vertex2 = map(int, input().split()) #–í–≤–æ–¥ –Ω–µ–Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã—Ö —Ä—ë–±–µ—Ä –≤–µ—Ä—à–∏–Ω
            self.neighbors[vertex1].append(vertex2) #–°–æ–∑–¥–∞–Ω–∏–µ  –≤–µ—Ä—à–∏–Ω
            self.neighbors[vertex2].append(vertex1)

    def isBipartite(self):  # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ –¥–≤—É–¥–æ–ª—å–Ω–æ—Å—Ç—å –≥—Ä–∞—Ñ–∞
        return all(self.isBipartiteComponent(vertex) for vertex in range(1, self.numVertices)
                   if self.groupOf(vertex) == None)

    def isBipartiteComponent(self, vertex):  # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ —á–∞—Å—Ç—å –¥–≤—É–¥–æ–ª—å–Ω–æ–≥–æ –≥—Ä–∞—Ñ–∞
        if len(self.neighbors[vertex]) == 0:
            return True

        self.groups[0].add(vertex) # –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –≤–µ—Ä—à–∏–Ω –≤ –≥—Ä—É–ø–ø—É
        queue = [vertex]

        while queue:
            vertex = queue.pop() # –≤–æ–∑–≤—Ä–∞—â–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ –æ—á–µ—Ä–µ–¥–∏
            groupOfVertex = self.groupOf(vertex) 

            for neighbor in self.neighbors[vertex]:
                groupOfNeighbor = self.groupOf(neighbor)

                if groupOfNeighbor == groupOfVertex: #–ï—Å–ª–∏ –≥—Ä—É–ø–ø–∞ –≤–µ—Ä—à–∏–Ω —Ä–∞–≤–Ω–∞ –≥—Ä—É–ø–ø–µ –Ω–µ–Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã—Ö —Ä—ë–±–µ—Ä, –≤–µ—Ä–Ω—ë–º false
                    return False

                elif groupOfNeighbor == None: #–æ—á–∏—â–∞–µ–º
                    self.groups[not groupOfVertex].add(neighbor)
                    queue.append(neighbor)

        return True

    def groupOf(self, vertex): #—Ñ—É–Ω–∫—Ü–∏—è –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∏ –Ω–µ–Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã—Ö —Ä—ë–±–µ—Ä
        return 0 if vertex in self.groups[0] else 1 if vertex in self.groups[1] else None


def solve():
    ans = 0
    graph = Graph() #–∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –≥—Ä–∞—Ñ–∞
    if graph.isBipartite(): #–ø—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ –¥–≤—É–¥–æ–ª—å–Ω–æ—Å—Ç—å –≥—Ä–∞—Ñ–∞
        for group in graph.groups: #–≤—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –Ω–∞ —ç–∫—Ä–∞–Ω
            count_vertex = len(group)
            #print(count_vertex)
            index_of_vertex = " ".join(map(str, group))
            #print(index_of_vertex)
    else:
        ans = -1
        
    return count_vertex,index_of_vertex,ans

ans = 0
count_vertex, index_of_vertex, ans = solve()
if (ans == -1):
    print("–ù–µ–≤–æ–∑–º–æ–∂–Ω–æ –ø–æ–¥–µ–ª–∏—Ç—å –≥—Ä–∞—Ñ—ã")
else:
    print("–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:")
    print(count_vertex)
    print(index_of_vertex)
```

```python

def check_type_int (n: int):
    if not isinstance(n, int): # –ü—Ä–æ–≤–µ—Ä–∫–∞ n –Ω–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ —Ç–∏–ø—É –¥–∞–Ω–Ω—ã—Ö int
        raise TypeError("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥—Ä—É–∑–µ–π –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å —Ü–µ–ª—ã–º —á–∏—Å–ª–æ–º") # –ü–æ–¥–Ω–∏–º–∞–µ—Ç –∏—Å–∫–ª—é—á–µ–Ω–∏–µ TypeError

def check_type_int_for_elems_in_list(n: list):
    for elems in range(len(n)):
        if not isinstance(n[elems], int): #–ü—Ä–æ–≤–µ—Ä–∫–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å–ø–∏—Å–∫–∞ –Ω–∞ —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö Int
            raise TypeError("–ß–∏—Å–ª–∞ –Ω–∞ —Ñ—É—Ç–±–æ–ª–∫–µ –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å —Ü–µ–ª—ã–º–∏ —á–∏—Å–ª–∞–º–∏") # –í—ã–∑—ã–≤–∞–µ—Ç –∏—Å–∫–ª—é—á–µ–Ω–∏–µ TypeError

def check_value(n: int, left_edge: int, right_edge: int):
    if not (left_edge <= n <= right_edge): # –ü—Ä–æ–≤–µ—Ä–∫–∞ n –Ω–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ –≤—Ö–æ–∂–¥–µ–Ω–∏—è –≤ –≥—Ä–∞–Ω–∏—Ü—ã –¥–æ–ø—É—Å—Ç–∏–º—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
        raise ValueError("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥—Ä—É–∑–µ–π –¥–æ–ª–∂–Ω–æ –ø—Ä–∏–Ω–∏–º–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ –æ—Ç 1 –¥–æ 50") # –ü–æ–¥–Ω–∏–º–∞–µ—Ç –∏—Å–∫–ª—é—á–µ–Ω–∏–µ ValueError

def check_value_in_list(n: list, left_edge: int, right_edge: int):
    for elems in range(len(n)):
        if not (left_edge <= n[elems] <= right_edge): #–ü—Ä–æ–≤–µ—Ä–∫–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å–ø–∏—Å–∫–∞ –Ω–∞ –≤—ã—Ö–æ–¥ –∏–∑ –≥—Ä–∞–Ω–∏—Ü –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–π
            raise ValueError("–ß–∏—Å–ª–∞ –Ω–∞ —Ñ—É—Ç–±–æ–ª–∫–µ –¥–æ–ª–∂–Ω—ã –ø—Ä–∏–Ω–∏–º–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ –æ—Ç 1 –¥–æ 105") #–í—ã–∑–æ–≤ –∏—Å–∫–ª—é—á–µ–Ω–∏—è ValueError

def solution(a: list,t: int):
    for _ in range(t):
        if sorted(a[::2])==sorted(a)[::2]: #–ï—Å–ª–∏ –≤ —Å–æ—Ä—Ç–∏—Ä—É–µ–º–æ–º —Å–ø–∏—Å–∫–µ, –∫–∞–∂–¥—ã–π –≤—Ç–æ—Ä–æ–π —ç–ª–µ–º–µ–Ω—Ç —Ä–∞–≤–µ–Ω –∫–∞–∂–¥–æ–º—É –≤—Ç–æ—Ä–æ–º—É 
            return True                    #—ç–ª–µ–º–µ–Ω—Ç—É –≤ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–º —Å–ø–∏—Å–∫–µ, –≤–µ—Ä–Ω—É—Ç—å true
        else:
            return False


if __name__ == '__main__':
    print("–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:\n –í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö")
    t = int(input()) # –í–≤–æ–¥ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö
    check_type_int(t) # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ —Ç–∏–ø int
    check_value(t,1,50) #–ü—Ä–æ–≤–µ—Ä–∫–∞ –≤—ã—Ö–æ–¥–∞ –∑–∞ –≥—Ä–∞–Ω–∏—Ü—É –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–π
    for _ in range(t):
        print("–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥—Ä—É–∑–µ–π:")
        n = int(input()) #–í–≤–æ–¥ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –¥—Ä—É–∑–µ–π 
        check_type_int(n) 
        check_value(n,1,105)
        print("–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–∞ –Ω–∞ —Ñ—É—Ç–±–æ–ª–∫–µ:")
        a = list(map(int,input().split())) # –í–≤–æ–¥ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —á–∏—Å–µ–ª –Ω–∞ —Ñ—É—Ç–±–æ–ª–∫–µ
        check_type_int_for_elems_in_list(a) #–ü—Ä–æ–≤–µ—Ä–∫–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å–ø–∏—Å–∫–∞ –Ω–∞ —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö int
        check_value_in_list(a,1,105) #–ü—Ä–æ–≤–µ—Ä–∫–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å–ø–∏—Å–∫–∞ –Ω–∞ –≤—ã—Ö–æ–¥ –∏–∑ –≥—Ä–∞–Ω–∏—Ü –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–π
        if (solution(a,t)): # –ï—Å–ª–∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ true, —Ç–æ –≤—ã–≤–æ–¥–∏–º –¥–∞, –∏–Ω–∞—á–µ –Ω–µ—Ç
            print('YES')
        else:
            print('NO')
```

```python
def check_type_int (n: int):
    if not isinstance(n, int): # –ü—Ä–æ–≤–µ—Ä–∫–∞ n –Ω–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ —Ç–∏–ø—É –¥–∞–Ω–Ω—ã—Ö int
        raise TypeError("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–æ–∫—Å—ë—Ä–æ–≤ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å —Ü–µ–ª—ã–º —á–∏—Å–ª–æ–º") # –ü–æ–¥–Ω–∏–º–∞–µ—Ç –∏—Å–∫–ª—é—á–µ–Ω–∏–µ TypeError

def check_value(n: int, left_edge: int, right_edge: int):
    if not (left_edge <= n <= right_edge): # –ü—Ä–æ–≤–µ—Ä–∫–∞ n –Ω–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ –≤—Ö–æ–∂–¥–µ–Ω–∏—è –≤ –≥—Ä–∞–Ω–∏—Ü—ã –¥–æ–ø—É—Å—Ç–∏–º—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
        raise ValueError("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–æ–∫—Å—ë—Ä–æ–≤ –¥–æ–ª–∂–Ω–æ –ø—Ä–∏–Ω–∏–º–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ –æ—Ç 1 –¥–æ 150000") # –ü–æ–¥–Ω–∏–º–∞–µ—Ç –∏—Å–∫–ª—é—á–µ–Ω–∏–µ ValueError

def solution(l: int, r: int, m: int, res: int, res2: int):
    for i in range(l, r+1):
        n = res2 // i #–ù–∞—Ö–æ–¥–∏–º —Å–µ—Ä–µ–¥–∏–Ω—É
        if (n* i < res or n*1>res2): #–ï—Å–ª–∏ –Ω–µ –≤—ã—Ö–æ–¥–∏—Ç –∑–∞ —Ä–∞–º–∫–∏, –ø—Ä–æ–¥–æ–ª–∂–∞–µ–º
            continue
        flag = n * i - m  #–í—ã—Å—á–∏—Ç—ã–≤–∞–µ–º —Ñ–ª–∞–≥
        if flag >= 0:
            return (i,l,l+flag) #–í–æ–∑–≤—Ä–∞—â–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏—è —Ç—Ä–æ–π–∫–∏ —á–∏—Å–µ–ª
        else:
            return (i,r,r+flag) #–ï—Å–ª–∏ —É—Å–ª–æ–≤–∏–µ –Ω–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è –≤–æ–∑–≤—Ä–∞—â–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –∏ –≤—ã—Ö–æ–¥–∏–º –∏–∑ —Ü–∏–∫–ª–∞
            break

if __name__ == '__main__':
    print("–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: \n–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö")
    t = int(input()) # –í–≤–æ–¥ –Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö
    check_type_int(t) #–ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ —Ç–∏–ø int
    check_value(t,1,20) # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ –≤—ã—Ö–æ–¥ –∏–∑ –≥—Ä–∞–Ω–∏—Ü –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–π
    for _ in range(t):
        print("–í–≤–µ–¥–∏—Ç–µ —Ç—Ä–æ–π–∫—É —á–∏—Å–µ–ª:")
        l, r, m = map(int, input().strip().split(" ")) #–í–≤–æ–¥ 3–∫–∏ —á–∏—Å–µ–ª
        res1 = m - r + l #–†–∞—Å—Å—á—ë—Ç –ø–æ –¥–∞–Ω–Ω–æ–π –≤ —É—Å–ª–æ–≤–∏–∏ —Ñ–æ—Ä–º—É–ª–µ
        res2 = m + r - l
        ans1,ans2,ans3 = solution(l,r,m,res1,res2) #–ü–æ–ª—É—á–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –¥–ª—è –æ—Ç–≤–µ—Ç–∞
        print("–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:")
        print(ans1,ans2,ans3)
```

```python

def check_type_int (n: int):
    if not isinstance(n, int): # –ü—Ä–æ–≤–µ—Ä–∫–∞ n –Ω–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ —Ç–∏–ø—É –¥–∞–Ω–Ω—ã—Ö int
        raise TypeError("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–æ–∫—Å—ë—Ä–æ–≤ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å —Ü–µ–ª—ã–º —á–∏—Å–ª–æ–º") # –ü–æ–¥–Ω–∏–º–∞–µ—Ç –∏—Å–∫–ª—é—á–µ–Ω–∏–µ TypeError

def check_value(n: int, left_edge: int, right_edge: int):
    if not (left_edge <= n <= right_edge): # –ü—Ä–æ–≤–µ—Ä–∫–∞ n –Ω–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ –≤—Ö–æ–∂–¥–µ–Ω–∏—è –≤ –≥—Ä–∞–Ω–∏—Ü—ã –¥–æ–ø—É—Å—Ç–∏–º—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
        raise ValueError("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–æ–∫—Å—ë—Ä–æ–≤ –¥–æ–ª–∂–Ω–æ –ø—Ä–∏–Ω–∏–º–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ –æ—Ç 1 –¥–æ 150000") # –ü–æ–¥–Ω–∏–º–∞–µ—Ç –∏—Å–∫–ª—é—á–µ–Ω–∏–µ ValueError

def check_type_int_for_elems_in_list(n: list):
    for elems in range(len(n)):
        if not isinstance(n[elems], int): #–ü—Ä–æ–≤–µ—Ä–∫–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å–ø–∏—Å–∫–∞ –Ω–∞ —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö Int
            raise TypeError("–í–µ—Å –±–æ–∫—Å—ë—Ä–∞ –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å —Ü–µ–ª—ã–º —á–∏—Å–ª–æ–º") # –í—ã–∑—ã–≤–∞–µ—Ç –∏—Å–∫–ª—é—á–µ–Ω–∏–µ TypeError

def check_value_in_list(n: list, left_edge: int, right_edge: int,power_for_right_edge: int):
    for elems in range(len(n)):
        if not (left_edge <= n[elems] <= math.pow(right_edge,power_for_right_edge)): #–ü—Ä–æ–≤–µ—Ä–∫–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å–ø–∏—Å–∫–∞ –Ω–∞ –≤—ã—Ö–æ–¥ –∏–∑ –≥—Ä–∞–Ω–∏—Ü –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–π
            raise ValueError("–í–µ—Å –±–æ–∫—Å—ë—Ä–∞ –¥–æ–ª–∂–µ–Ω –ø—Ä–∏–Ω–∏–º–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ –æ—Ç 1 –¥–æ 150000") #–í—ã–∑–æ

def solution(t: int,a:list,b:list ):
    n=math.floor((-1+math.sqrt(1+8*(b-a)))/2) #–í—ã—Å—á–∏—Ç—ã–≤–∞–µ–º –Ω–∞–∏–±–æ–ª—å—à–µ–µ —á–∏—Å–ª–æ
    a=a+n*(n+1)/2 #–í—ã—Å—á–∏—Ç—ã–≤–∞–µ–º –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –∞
    while ((a-b)%2)!=0 or a<b: # –ü–æ–∫–∞ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è —É—Å–ª–æ–≤–∏–µ
         n+=1 # n - —Å—á—ë—Ç—á–∏–∫
         a+=n #–ü—Ä–∏–±–∞–≤–ª—è–µ–º –∫ –∞ n
    return n

if __name__ =='__main__':
    print("–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: \n–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–æ—Ä–æ–≤ –¥–∞–Ω–Ω—ã—Ö")
    temp = int(input()) #–í–≤–æ–¥–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–æ—Ä–æ–≤ –¥–∞–Ω–Ω—ã–∑
    check_type_int(temp) # –ü—Ä–æ–≤–µ—Ä—è–µ–º –Ω–∞ int –∑–Ω–∞—á–µ–Ω–∏–µ
    check_value(temp,1,100) #–ü—Ä–æ–≤–µ—Ä—è–µ–º –Ω–∞ –≤—ã—Ö–æ–¥ –∑–∞ —Ä–∞–º–∫–∏ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–π
    for _ in range(temp):
        print("–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ –∞ –∏ b:")
        [a, b] = sorted(list(map(int, input().split()))) #–í–≤–æ–¥–∏–º –∏ —Å–æ—Ä—Ç–∏—Ä—É–µ–º —Å–ø–∏—Å–æ–∫ –∏–∑ int —ç–ª–µ–º–µ–≥—Ç–æ–≤
        n = solution(temp,a,b) #–í –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é n –≤–Ω–æ—Å–∏–º –∑–Ω–∞—á–µ–Ω–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ–º—ã–µ –∏–∑ —Ñ—É–Ω–∫—Ü–∏–∏
        print("–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:")
        print(n)
```
